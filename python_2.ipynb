### Задание 3
Список поисковых запросов. Получить распределение количества слов в них. Т. е. поисковых запросов из одного слова 5%, из двух - 7%, из трех - 3% итд.
queries = [
	'смотреть сериалы онлайн',
	'новости спорта',
	'афиша кино',
	'курс доллара',
	'сериалы этим летом',
	'курс по питону',
	'сериалы про спорт',
]

```python
queries = [
	'смотреть сериалы онлайн',
	'новости спорта',
	'афиша кино',
	'курс доллара',
	'сериалы этим летом',
	'курс по питону',
	'сериалы про спорт',
]
SearchPhrases = len(queries)
OneWords,TwoWords,ThreeWords = 0,0,0 #0 пробелов

for i in queries:
    if i.count(' ') == 0:
        OneWords += 1
    elif i.count(' ') == 1:
        TwoWords += 1
    elif i.count(' ') ==2:
        ThreeWords += 1
        
print('Поисковых запросов из одного слова {:.0%}'.format(OneWords/SearchPhrases))
print('Поисковых запросов из двух слов {:.0%}'.format(TwoWords/SearchPhrases))
print('Поисковых запросов из трех слов {:.0%}'.format(ThreeWords/SearchPhrases))
```
